```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Loading in data
PCOS_inf = pd.read_csv('/content/drive/MyDrive/College/3rd Year [2023-2024]/Spring 2024/Data Science Bootcamp (S24)/Capstone/PCOS_infertility.csv')
PCOS_woinf = pd.read_excel('/content/drive/MyDrive/College/3rd Year [2023-2024]/Spring 2024/Data Science Bootcamp (S24)/Capstone/PCOS_data_without_infertility.xlsx','Full_new')

#Checking column names
print("Columns in PCOS_inf:", PCOS_inf.columns)
print("Columns in PCOS_woinf:", PCOS_woinf.columns)

#Merging the data files
df = pd.merge(PCOS_woinf, PCOS_inf, on='Patient File No.', suffixes=('', '_wo'), how='left')

#Dropping repeated/unnecessary columns
df = df.drop(['Unnamed: 44','Sl. No_wo', 'PCOS (Y/N)_wo', '  I   beta-HCG(mIU/mL)_wo','II    beta-HCG(mIU/mL)_wo', 'AMH(ng/mL)_wo'], axis=1)

#Renaming column due to misspelling in original df
df.rename(columns={'Marraige Status (Yrs)': 'Marriage Status (Yrs)'}, inplace=True)

#Preview of the merged data
print(df.head())
print("Columns in df:", df.columns)



```
